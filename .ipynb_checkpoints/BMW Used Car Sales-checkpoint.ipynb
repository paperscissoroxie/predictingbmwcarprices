# Modeling

## Linear Regression

```python
# Preparing data for modeling

# Create DataFrame with features only
bmw_used_car_sales_f = bmw_used_car_sales.copy()
bmw_used_car_sales_f.drop(['price'], inplace = True, axis = 1)

# Segregate features and labels into separate variables
X = bmw_used_car_sales_f
y = bmw_used_car_sales['price']

print('Number of independent variables: ' + str(X.shape[1]))
print()

from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV

# Standardize features
std = StandardScaler()
X = std.fit_transform(X)

# Partition the dataset in train + validation sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

print('X_train : ' + str(X_train.shape))
print('X_test : ' + str(X_test.shape))
print('y_train : ' + str(y_train.shape))
print('y_test : ' + str(y_test.shape))
```

```python
# RMSE
def rmse_cv_train(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv = 5))
    return(rmse)

def rmse_cv_test(model):
    rmse= np.sqrt(-cross_val_score(model, X_test, y_test, scoring='neg_mean_squared_error', cv = 5))
    return(rmse)
```

```python
# Linear Regression

# Training the algorithm
lr = LinearRegression().fit(X_train, y_train)

# Look at predictions on training and validation set
print('RMSE on Train set : {0:0.3f}'.format(rmse_cv_train(lr).mean()))
print('RMSE on Test set : {0:0.3f}'.format(rmse_cv_test(lr).mean()))

# Print the coefficient of determination
r_sq = lr.score(X_train, y_train)
print('\nThe coefficient of determination calculated by the linear regression : {0:0.2f}'.format(r_sq))

# Print intercept and coefficients
lr_int = lr.intercept_
print('The intercept calculated by the linear regression : {0:0.2f}'.format(lr_int))
lr_coef = pd.DataFrame(lr.coef_, bmw_used_car_sales_f.columns, columns=['Coefficient'])
print('The coefficients calculated by the linear regression :\n\n', lr_coef)
#print('The coefficients calculated by the linear regression :', (*['{0:0.2f}'.format(i) for i in lr_coef]))

y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)

# Plot residuals
plt.scatter(y_train_pred, y_train_pred - y_train, c = 'blue', marker = 's', label = 'Train data')
plt.scatter(y_test_pred, y_test_pred - y_test, c = 'lightgreen', marker = 's', label = 'Validation data')
plt.title('Linear Regression')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.legend(loc = 'lower right')
plt.hlines(y = 0, xmin = 10.5, xmax = 13.5, color = 'red')
plt.show()

# Plot predictions
plt.scatter(y_train_pred, y_train, c = 'blue', marker = 's', label = 'Train data')
plt.scatter(y_test_pred, y_test, c = 'lightgreen', marker = 's', label = 'Validation data')
plt.title('Linear Regression')
plt.xlabel('Predicted Values')
plt.ylabel('Real Values')
plt.legend(loc = 'lower right')
plt.plot([10.5, 13.5], [10.5, 13.5], c = 'red')
plt.show()
```

```python
# Make some predictions

y_pred = lr.predict(X_test)
```

```python
# Compare actual output values for 'X_test' with predicted values

act_v_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
act_v_pred
```

The model was used to predict new values and compared to actual values. The predicted and actual values are close to each other. Further assessment of the evaluation of the model performance is produced below.

```python
# Evaluate algorithm
from sklearn import metrics

print('Mean Absolute Error (MAE): {0:0.2f}'.format(metrics.mean_absolute_error(y_test, y_pred)))
print('Mean Squared Error (MSE): {0:0.2f}'.format(metrics.mean_squared_error(y_test, y_pred)))
print('Root Mean Squared Error (RMSE): {0:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
mape = np.mean(np.abs((y_test - y_pred) / np.abs(y_test)))
print('Mean Absolute Percentage Error (MAPE): {0:0.2f}'.format(round(mape * 100, 2)))
print('Accuracy:', round(100*(1 - mape), 2))

bmw_used_car_sales['price'].describe().round(2)
```

The value of the RMSE is 0.20, less than 10% of the mean value of the price, 9.92, which shows that the model is performing well.

Log transformation was performed on the data used to train the model. Back transformation must be performed on the predicted values to express them on the same scale as the original data by applying the inverse of the original transformation.

```python
# Back transform
y_pred = np.exp(y_pred)
y_test = np.exp(y_test)
```

## Random Forest

```python
# Preparing data for modeling

# Create DataFrame with features only
bmw_used_car_sales_f = bmw_used_car_sales.copy()
bmw_used_car_sales_f.drop(['price'], inplace = True, axis = 1)

# Segregate features and labels into separate variables
X = bmw_used_car_sales_f
y = bmw_used_car_sales['price']

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=42)
```

```python
# Random forest regressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rf = RandomForestRegressor()

param_grid = { "criterion" : ["mse"]
              , "min_samples_leaf" : [1,5,1]
              , "min_samples_split" : [1,5,1]
              , "max_depth": [10]
              , "n_estimators": [500]}

gs = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
gs = gs.fit(X_train, y_train)
```

```python
# Performance evaluation

print('Best Score: %.3f' % gs.best_score_)
print('Best Parameters: ', gs.best_params_)
```

```python
bp = gs.best_params_
forest = RandomForestRegressor(criterion=bp['criterion'],
                              min_samples_leaf=bp['min_samples_leaf'],
                              min_samples_split=bp['min_samples_split'],
                              max_depth=bp['max_depth'],
                              n_estimators=bp['n_estimators'])
forest.fit(X_train, y_train)

print('Score: %.3f' % forest.score(X_test, y_test))

# Look at predictions on training and validation set
print('RMSE on Train set : {0:0.3f}'.format(rmse_cv_train(forest).mean()))
print('RMSE on Test set : {0:0.3f}'.format(rmse_cv_test(forest).mean()))
```

```python
# Make predictions for the test set

y_pred = forest.predict(X_test)
```

```python
# Model Accuracy

print('Mean Absolute Error (MAE): {0:0.2f}'.format(metrics.mean_absolute_error(y_test, y_pred)))
print('Mean Squared Error (MSE): {0:0.2f}'.format(metrics.mean_squared_error(y_test, y_pred)))
print('Root Mean Squared Error (RMSE): {0:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
mape = np.mean(np.abs((y_test - y_pred) / np.abs(y_test)))
print('Mean Absolute Percentage Error (MAPE): {0:0.2f}'.format(round(mape * 100, 2)))
print('Accuracy:', round(100*(1 - mape), 2))

bmw_used_car_sales['price'].describe().round(2)
```

## Recommendation Engine

```python
# Imports
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from matplotlib.pyplot import imshow
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, zoomed_inset_axes
import _pickle as cPickle
import gzip
```

```python
def load_zipped_pickle(filename):
    with gzip.open(filename, 'rb') as f:
        loaded_object = cPickle.load(f)
        return loaded_object

def save_zipped_pickle(obj, filename, protocol=-1):
    with gzip.open(filename, 'wb') as f:
        cPickle.dump(obj, f, protocol)
```

```python
# DatatFrame with descriptions
bmw_used_car_sales_desc = pd.read_csv('data/bmw.csv')

print(bmw_used_car_sales_desc.head())
```

```python
def create_description(df):
    m = str(df.model)
    y = str(df.year)
    t = str(df.transmission)
    f = str(df.fuelType)
    e = str(df.engineSize)

    p = str(df.price)
    mile = str(df.mileage)
    mpg = str(df.mpg)
    
    s = " ".join([m, y, t, f, e])
    return s
    
bmw_used_car_sales_desc['description'] = bmw_used_car_sales_desc.apply(create_description, axis=1)
```

```python
print(bmw_used_car_sales_desc.description.loc[5], '\n')
print(bmw_used_car_sales_desc.description.loc[10780], '\n')
print(bmw_used_car_sales_desc.description.head())
```

```python
# Tokenization using TfidfVectorizer
tf_vectorizer = TfidfVectorizer(ngram_range=(1,4),
                               min_df = 0)

tfidf_matrix = tf_vectorizer.fit_transform(bmw_used_car_sales_desc['description'])
```

```python
# Compute the cosine similarity for each listing against all other listings
from datetime import datetime

d1 = datetime.now()
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
d2 = datetime.now()
time_diff_1 = int((d2-d1).total_seconds())

#print('This took %s seconds' % (time_diff_1))

def time_diff(d1, d2):
    n = (int((d2-d1).total_seconds()))
    if n == 1:
        print('This took %s second' % n)
    else:
        print('This took %s seconds' % n)

time_diff(d1, d2)
```

Using the consine similarity results, create two additional columns:

1. similar_all: the 20 most similar results
2. similar_other_model: the 20 most similar results that are different models

```python
d1 = datetime.now()
all_similar = []
other_model_similar = []

for idx, row in bmw_used_car_sales_desc.iterrows():
    if (idx % 5000) == 0:
        print("at index %s" % idx)
        
    # Limit to the 200 'most similar' listings
    similar_indices = cosine_similarities[idx].argsort()[:-200:-1]
    
    # Remove the parent item from the list, so
    # that it doesn't show up due to the presence
    # of several indexes with the same value
    similar_indices = similar_indices[similar_indices != idx]

    
    # Return the similarity and the index of the similar listing
    similar_items = [(cosine_similarities[idx][i], bmw_used_car_sales_desc.index[i])
                     for i in similar_indices]
    
    # Keep the list of all similar results
    # limit to 20 items
    all_similar.append(similar_items[0:20])
    
    # Look for other models
    sim_diff_model = []
    for position, (sim, item) in enumerate(similar_items):

        # Get 20 similar listings that are a different make
        if bmw_used_car_sales_desc.model.loc[idx] != bmw_used_car_sales_desc.model.loc[item]:
            if len(sim_diff_model) <= 20:
                sim_diff_model.append(similar_items[position])
             
    # Append the other similar makes
    other_model_similar.append(sim_diff_model)

    
bmw_used_car_sales_desc['similar_all'] = all_similar
bmw_used_car_sales_desc['similar_other_model'] = other_model_similar

# Save the results
#bmw_sim_file = 'output/bmw_used_car_sales_probs_and_similarities.pgz'
#save_zipped_pickle(bmw_used_car_sales_desc, bmw_sim_file)

d2 = datetime.now()
time_diff_1 = int((d2-d1).total_seconds())

#print('This took %s seconds' % (time_diff_1))

def time_diff(d1, d2):
    n = (int((d2-d1).total_seconds()))
    if n == 1:
        print('This took %s second' % n)
    else:
        print('This took %s seconds' % n)

time_diff(d1, d2)
```

```python
# Inspecting output
x = 50

print('Parent listing: %s\n' % bmw_used_car_sales_desc.description.loc[x])
print('Similarity score of top-five most similar listings:')
for i in bmw_used_car_sales_desc.similar_all.loc[x][0:5]:
        print(i[0], bmw_used_car_sales_desc.description.loc[i[1]])
```

### Create functions to:

#### 1. Return info about any listing

```python
def get_info_to_display(ad_index):
    # Take an index from bmw_used_car_sales_desc and return a list with details for printing
    
    # Specifics from the listing
    year = str(bmw_used_car_sales_desc.year.loc[ad_index])
    model = str(bmw_used_car_sales_desc.model.loc[ad_index])
    price = int(bmw_used_car_sales_desc.price.loc[ad_index])
    miles = int(bmw_used_car_sales_desc.mileage.loc[ad_index])
    
    # Create a list for specifics
    display_info = [model, year, miles, price]
    
    return display_info
```

#### 2. Return similar model with a price 15% lower than the original listing

```python
def get_lower_price(ad_index):
    parent = ad_index
    parent_price = bmw_used_car_sales_desc.price.loc[parent]
    # Return a listing that has a vehicle similar to the original listing,
    # but with at least a 15% lower price
    
    # Take the parent and list of related vehicles for any of the
    # defined searches and return either the lower priced index, or None 
    def get_lower_price_index(parent, related_list):

        for item_score, item_index in related_list:
            # Find listing with lower price
            d = bmw_used_car_sales_desc.price.loc[parent] - bmw_used_car_sales_desc.price.loc[item_index]        
            # Look for price 15% lower than parent
            if (d > (parent_price * 0.15)):
                return item_index

    # Look for something that is in the similar overall list
    similar_all_list = bmw_used_car_sales_desc.similar_all.loc[parent]
    if len(similar_all_list) > 0:
        lpi = get_lower_price_index(parent, similar_all_list)
        # Check to see if a value was returned or not
        if lpi:
            # If yes, return the details
            return get_info_to_display(lpi)
        
    # If no, then look for something that is similar and a different model
    sim_list = bmw_used_car_sales_desc.similar_other_model.loc[parent]
    if len(sim_list) > 0:
        lpi = get_lower_price_index(parent, sim_list)
        # Check to see if a value was returned or not
        if lpi:
            return get_info_to_display(lpi)
    
    # If none of the above, return None
    return None
```

#### 3. Return similar model with at least 10% fewer miles than the parent listing.

```python
def get_lower_miles(ad_index):
    # Return a listing that has a vehicle similar to the original listing,
    # but with at least 10% fewer miles
    parent = ad_index
    parent_miles = bmw_used_car_sales_desc.mileage.loc[parent]
    
    # Take the parent and list of related vehicles for any of the
    # defined searches and return either the index, or None
    def get_lower_miles_index(parent, related_list):

        for item_score, item_index in related_list:
            # Find listing with lower miles
            d = parent_miles - bmw_used_car_sales_desc.mileage.loc[item_index]        
            # Look for 10% lower miles than parent
            if (d > (parent_miles * 0.1)):
                return item_index

    # Look for something that is in the similar overall list
    similar_all_list = bmw_used_car_sales_desc.similar_all.loc[parent]
    if len(similar_all_list) > 0:
        lpi = get_lower_miles_index(parent, similar_all_list)      
        # Check to see if a value was returned or not
        if lpi:
            # If yes
            return get_info_to_display(lpi)
    
    # If no, then look for something that is similar and a different model
    sim_list = bmw_used_car_sales_desc.similar_other_model.loc[parent]
    if len(sim_list) > 0:
        lpi = get_lower_miles_index(parent, sim_list)    
        # Check to see if a value was returned or not
        if lpi:
            return get_info_to_display(lpi)
    
    
    # If none of the above, return None
    return None
```

#### 4. Return similar model with a price around that is 1-30% higher than the parent.

```python
def get_stretch_goal(ad_index):
    # This should return a listing that is slightly 'nicer' than
    # the parent listing, or at least more expensive
    parent = ad_index
    parent_price = bmw_used_car_sales_desc.price.loc[parent]
    parent_miles = bmw_used_car_sales_desc.mileage.loc[parent]
    
    def get_higher_price(parent, related_list):

        for item_score, item_index in related_list:
            # Find listing with higher price
            d = bmw_used_car_sales_desc.price.loc[item_index] - bmw_used_car_sales_desc.price.loc[parent]        
            
            # Look for price 1-30% higher than parent
            if (d > parent_price * .01) & (d <= (parent_price * .30)):
                if (item_index != get_lower_miles(parent)[3]) & (item_index != get_lower_miles(parent)[3]):
                    return item_index

    # Look for something similar 
    similar_all_list = bmw_used_car_sales_desc.similar_all.loc[parent]
    if len(similar_all_list) > 0:
        lpi = get_higher_price(parent, similar_all_list)
        # Check to see if a value was returned or not
        if lpi:
            # If yes
            return get_info_to_display(lpi)
        
    # If no, look for anything similar
    similar_other_list = bmw_used_car_sales_desc.similar_other_make.loc[parent]
    if len(similar_other_list) > 0:
        lpi = get_higher_price(parent, similar_other_list)
        # Check to see if a value was returned or not
        if lpi:
            return get_info_to_display(lpi)

    return None
```
